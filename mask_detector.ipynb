```python
import pyfirmata
import time
import tensorflow as tf
import cv2
import os
# import matplotlib.pyplot as plt
import numpy as np
```

```python
new_model = tf.keras.models.load_model('Untitled/by_Ankush.h5')
# new_model = tf.keras.models.load_model('my_model.h5')
```

```python
path = "haarcascade_frontalface_default.xml"
capture = cv2.VideoCapture(0)
data = []
font = cv2.FONT_HERSHEY_COMPLEX

while True:
    flag, img = capture.read()
    cv2.rectangle(img,(0,0), (100,100), (0,0,0), thickness=-1) # (-1 and cv.filled both works same)
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + path)

    if flag:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        #cv.imshow('Gray', img)
        faces = faceCascade.detectMultiScale(gray,1.1,4)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
            face = img[y:y+h, x:x+h, :]
            face = cv2.resize(face ,(224,224))
            face = np.expand_dims(face,axis = 0)
            face = face/255.0
            font = cv2.FONT_HERSHEY_SIMPLEX
            pred = new_model.predict(face)
            
            font_scale = 1.5
            font = cv2.FONT_HERSHEY_PLAIN
#             cv.putText(gray, n, (x,y), font,1,(255,255,255),2)
            
            if pred==1:
                n = "No Mask"
                cv2.circle(img,(50,50),40 ,(0,0,255),thickness=-1)
                cv2.line(img,(25,25), (75,75), (255,255,255), thickness=6) # (-1 and cv.filled both works same)
                cv2.line(img,(25,75), (75,25), (255,255,255), thickness=6)
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
                print(n)
                redLed.write(1)
                yellowLed.write(0)
                buzzer.write(1)
                #speak(n)
            else:
                n = "Mask"
                cv2.line(img,(15,50), (35,70), (0,255,0), thickness=10) # (-1 and cv.filled both works same)
                cv2.line(img,(32,70), (76,25), (0,255,0), thickness=10)
                cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
                print(n)
                redLed.write(0)
                yellowLed.write(1)
                buzzer.write(0)
#                 speak(n)
#             cv2.putText(img, n, (x,y), font,1,(255,255,255),2)
#             speak(n)
 
        cv2.imshow('Image',img)
        # 27 - Ascii value of Escape
        if cv2.waitKey(2) == 27:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
path = "haarcascade_frontalface_default.xml"
font_scale = 1.5
font = cv2.FONT_HERSHEY_PLAIN

#set the rectangle background to white
rectangle_bgr = (255,255,255)
#make a black image
img = np.zeros((500,500))
#set some text
text = "Some text in a box!"
#get the widht and height of the text box
(text_width,text_height) = cv2.getTextSize(text,font,fontScale = font_scale, thickness=1)[0]
# set the text start postion
text_offset_x = 10
text_offset_y = img.shape[0]-25
# make the coords of the box with asmall padding of two px
box_coords = ((text_offset_x,text_offset_y),(text_offset_x + text_width+2,text_offset_y - text_height - 2))
cv2.rectangle(img,box_coords[0],box_coords[1],rectangle_bgr,cv2.FILLED)
cv2.putText(img,text,(text_offset_x,text_offset_y),font,fontScale= font_scale,color=(0,0,0),thickness=1)


cap = cv2.VideoCapture(1)
#check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")
    
while True:
    ret,frame = cap.read()
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + path)
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    faces = faceCascade.detectMultiScale(gray,1.1,4)
    for x,y,w,h in faces:
        #roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        cv2.rectangle(frame,(x,y),(x+w, y+h),(255,0,0),2)
        facess = faceCascade.detectMultiScale(roi_color)
        if len(facess) == 0:
            face_roi = roi_color[ey:ey+eh, ex:ex+ew]
            print("face not detected")
        else:
            for (ex,ey,ew,eh) in facess:
                face_roi = roi_color[ey:ey+eh, ex:ex+ew]
            

    final_image = cv2.resize(face_roi ,(224,224))
    final_image = np.expand_dims(final_image,axis = 0)
    final_image = final_image/255.0
    font = cv2.FONT_HERSHEY_SIMPLEX
    predictions = new_model.predict(final_image)

    font_scale = 1.5
    font = cv2.FONT_HERSHEY_PLAIN
    
    
    if len(facess) == 0:
        status = "Face not detected"
        x1,y1,w1,h1 = 0,0,175,75
        # draw black background rectangle
        cv2.rectangle(frame,(x1,x1),(x1+w1,y1+h1),(0,0,0),-1)
        # add text
        cv2.putText(frame,status,(x1 + int(w1/10), y1 + int(h1/2)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,255,255),2)
        #cv2.putText(frame,status,(100, 150),font,3,(255,25,255),2,cv2.LINE_4)
        #cv2.rectangle(frame,(x,y),(x+w, y+h),(255,255,255))
        redLed.write(0)
        yellowLed.write(0)
        buzzer.write(0)
    
        
    elif (predictions==1):
#         time.sleep(0.1)
        status = "No_Mask"
        x1,y1,w1,h1 = 0,0,175,75
        # draw black background rectangle
        cv2.rectangle(frame,(x1,x1),(x1+w1,y1+h1),(0,0,0),-1)
        # add text
        cv2.putText(frame,status,(x1 + int(w1/10), y1 + int(h1/2)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,0,255),2)
        cv2.putText(frame,status,(100, 150),font,3,(0,0,255),2,cv2.LINE_4)
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,0,255))
        redLed.write(1)
        yellowLed.write(0)
        buzzer.write(1)
    
    else:
        status = "Face Mask"
        
        x1,y1,w1,h1 = 0,0,175,75
        # draw black background rectangle
        cv2.rectangle(frame,(x1,x1),(x1+w1,y1+h1),(0,0,0),-1)
        # add text
        cv2.putText(frame,status,(x1 + int(w1/10), y1 + int(h1/2)),cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,255,0),2)
        #cv2.putText(frame,status,(100, 150),font,3,(0,255,0),2,cv2.LINE_4)
        cv2.rectangle(frame,(x,y),(x+w, y+h),(0,255,0))
        redLed.write(0)
        yellowLed.write(1)
        buzzer.write(0)
#         time.sleep(0.1)
        
    
    
    cv2.imshow('Mask_detector', frame)
    
    if cv2.waitKey(2) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
